```python
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException, TimeoutException
from selenium import webdriver
import time
import pandas as pd
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
from selenium.webdriver.common.by import By

def get_reviews(num_fback, verbose):
    
    '''Gathers users' feedback as a dataframe'''
    
    #Initializing the webdriver
    options = webdriver.ChromeOptions()
    options.add_argument("start-maximized");
    options.add_argument("disable-infobars")
    options.add_argument("--disable-extensions")
    options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    
    #Change the path to where chromedriver is in your home folder.
    driver = webdriver.Chrome(executable_path="/Users/Navid/Documents/ds_salary_proj/chromedriver", options=options)

    url = 'https://play.google.com/store/apps/details?id=com.ncconsulting.skipthedishes_android&showAllReviews=true'
    driver.get(url)
    fbacks = []

    while len(fbacks) < num_fback:  #If true, should be still looking for new feedbacks.
#         time.sleep(1)

        reviews = WebDriverWait(driver, 3).until(ec.visibility_of_all_elements_located((By.XPATH, "//h3[.='User reviews']/following-sibling::div[1]/div")))
        for review in reviews:
            time.sleep(1)
            print("Progress: {}".format("" + str(len(fbacks)) + "/" + str(num_fback)))
            if len(fbacks) >= num_fback:
                break


            collected_successfully = False
            
          
            while not collected_successfully:
                try:               
                    full_review = WebDriverWait(review, 5).until(ec.element_to_be_clickable((By.XPATH, ".//button[text()='Full Review']")))
                    full_review.click()
                except TimeoutException:
                    pass
                   
                try: 
                    feedback = review.find_element_by_xpath(".//div[@jscontroller='LVJlx' and @class = 'UD7Dzf']").text
                    rating = review.find_element_by_xpath(".//div[@role='img']").get_attribute('aria-label')
                    collected_successfully = True
                except:
                    time.sleep(1)


            fbacks.append({'feedback' : feedback,
            "rating" : rating})

    return pd.DataFrame(fbacks)  #This line converts the dictionary object into a pandas DataFrame.
```

```python
df = get_reviewss(2000, True)
```
